```python
score = pd.DataFrame({'국어': [100, 80], '수학':[75, 90], 
                      '영어':[90, 95]}, index=['장화', '홍련'])
score
```

```python
stud = pd.DataFrame({'국어': [70, 85], '수학':[65, 100], 
                             '영어':[95, 65]}, index=['콩쥐', '팥쥐'])
score = score.append(stud)
score
```

```python
score2 = pd.concat([score, stud], axis=0)
score2 = score2.append(pd.Series({'국어':85, '수학':55, '영어':95}, name='해님'))
score2
```

```python
condition_list = [(score2['국어'] >= 90),
                 (score2['국어'] >= 80) & (score2['국어'] < 90),
                 (score2['국어'] >= 70) & (score2['국어'] < 80)]
choice_list = ['A', 'B', 'C']
```

```python
score2['점수'] = np.select(condition_list, choice_list, default='F')
score2
```

```python
health = pd.DataFrame({'연도':[2017, 2018, 2019, 2020, 2021, 2022], 
                       '키': [160, 162, 165, None, None, 166],
                       '몸무게': [53, 52, None, 50, 51, 54],
                       '시력': [1.2, None, 1.2, 1.2, 1.1, 0.8],
                       '병결': [None, None, None, 2, None, 1]})
health
```

```python
health.fillna(health.mean())
```

```python
health['병결'] = health['병결'].fillna(0)
health[['키', '몸무게']] = health[['키', '몸무게']].fillna(health[['키', '몸무게']].mean())
health['시력'] = health['시력'].fillna(health['시력'].mode()[0])
health
```
