```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdt
%matplotlib inline
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
%config InlineBackend.figure_format = 'svg'
```

```python
death = np.array([2, 1, 2, 4, 2, 5, 3, 3, 5, 6, 3, 8, 3,
                  3, 6, 3, 6, 5, 3, 5, 2, 6, 2, 3, 4, 3,
                  2, 9, 2, 2, 3, 2, 7, 3, 2, 10, 6, 2, 3,
                  1, 2, 3, 3, 4, 3, 2, 6, 2, 2, 3, 2, 3,
                  4, 3, 2, 3, 5, 2, 5, 5, 3, 4, 3, 6, 2,
                  1, 2, 3, 2, 6, 3, 3, 6, 3, 2, 3, 6, 4,
                  6, 5, 3, 5, 6, 2, 6, 3, 2, 3, 2, 6, 2,
                  6, 3, 3, 2, 6, 9, 6, 3, 6, 6, 2, 3, 2,
                  3, 5, 3, 5, 2, 3, 2, 3, 3, 1, 3, 3, 2,
                  3, 3, 4, 3, 6, 6, 3, 3, 3, 2, 3, 3, 6])
```

```python
table = (pd.Series(death)
.value_counts(sort=False)
.sort_index()
)
table.index = ["감염","각종암","순환기","호흡기","소화기",
               "사고사","비뇨기","정신병","노환","신경계"]
table
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 4))
# color 파라미터에 리스트를 전달하여 색상 여러개를 표시 가능
table.plot(kind='bar', rot=0, color=['black', 'blue'], ax=ax)
ax.tick_params(axis='both', labelsize=7)
ax.set_xlabel('사망원인')
ax.set_ylabel('빈도 수')
```

```python
# rc 모듈을 import 하여 글꼴 설정 가능
from matplotlib import rc
rc('font', family='Malgun Gothic')
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
table.plot(kind='pie', startangle=90, counterclock=False, autopct='%.1f%%', radius=0.8, ax=ax)
plt.rc('font', size=6)
```
